```python
import heapq
import time
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
class Node:
  def __init__(self, key, weights, h):
    self.key = key
    self.neighbors = None
    self.weights = weights
    self.f = 999
    self.g = 999
    self.h = h
```

```python
def weight(current, parent):
    return [current.weights[idx] for idx, n in enumerate(current.neighbors) if n.key == parent.key][0]
```

```python
def is_empty(Q):
  for q in Q:
    if q != []: return False
  return True
```

```python
def aStarAlgo_Parallel(start, target, k):
        Q = [[] for _ in range(k)]

        start.g = 0                 # calculate g value of start_node
        start.f = start.g + start.h # calculate f value of start_node

        H = {} # H, closed list
        H[start.key] = (start.f, start.g)
        
        heapq.heappush(Q[0], (start.f, start.key, start)) 
        m = None
        
        # parents contains an adjacency map of all nodes
        prev = {}
        prev[start] = start
                 
        while not is_empty(Q):
            S = []

            for i in range(k):
              if Q[i] == []:
                continue

              _, _, qi = heapq.heappop(Q[i])
              
              print(f"qi: {qi.key}")

              if qi.key == target.key:
                if m == None or qi.f < m.f: # compare f value
                  m = qi
              
                  # print()
                  # print(f"m: {m.key}")
                continue 
              
              # S <- S + Expand(qi)
              for n in qi.neighbors:
                if n.key != prev[qi].key and n.g > qi.g + weight(n, qi): # compare g value
                  n.g = qi.g + weight(n, qi)                             # update g value
                  n.f = n.g + n.h                                        # update f value
                  prev[n] = qi

                  # print(f"{n.key}.f: {n.f}")
                  # print(H)

                if n.key not in H or H[n.key][1] > n.g: # 이 코드를 논문에서는 T <- S 아래 부분에서 구현함. 뭐가 효율적인 건지?
                  S.append(n)
              
              print("S: ", end='')
              for s in S:    
                print(f"({s.key}, f:{s.f}, g:{s.g}) ", end='')
              print()
                  
                # print("prev: ", end='')
                # for p, v in prev.items():    
                #   print(f"{p.key}: {v.key} ", end='')
                # print()

            # end for 첫번째 블록
            
            if m != None:
              path = []
              distance = m.f
              while prev[m].key != m.key:
                path.append(m.key)
                m = prev[m]

              path.append(start.key)
              path.reverse()
              return path, distance
            
            # print("S: ", end='')
            # for s in S:    
            #   print(f"({s.key}, f:{s.f}, g:{s.g}) ", end='')
            # print()
            # print(f"H: {H}")

            cuckoo = CuckooHashing(len(S) * 2)
            for s_prime in S:
              cuckoo.insert(s_prime)

            T = cuckoo.getList()
            
            print("T: ", end='')
            for t in T:
              print(f"({t.key}, f:{t.f}, g:{t.g}) ", end='')
            print()

            for t_prime in T:
              i = random.randint(0,k-1)

              # print("t_prime.f:", t_prime.f)
              # print("i:", i)
              
              heapq.heappush(Q[i], (t_prime.f, t_prime.key, t_prime))              
              H[t_prime.key] = (t_prime.f, t_prime.g)
            
            # print(f"H: {H}")
            print("Q:", Q)
            # print()
```

```python
class CuckooHashing:
  def __init__(self, size):
    self.M = size
    self.T1 = [None] * size
    self.T2 = [None] * size
    self.N = 0
  
  def hash(self, key):       # 노드마다 키는 모두 다르므로 노드의 키를 아스키 코드로 바꿔서 해시값 계산에 사용
    return ord(key) % self.M # 이거는 엠엔 데이터에 따라서 달라질 수 있음
  
  def dhash(self, key):
    half = self.M // 2
    dh = ord(key) % half
    return (self.hash(key) + dh) % self.M
  
  def insert(self, node):
    newNode = node
    i = self.hash(newNode.key)
    currentTable = self.T1
    currentTable[i], oldNode = newNode, currentTable[i]

    while True:

      # print("T1:", self.T1)
      # print("T2:", self.T2)

      if oldNode == None or oldNode.key == newNode.key:
        return
        
      elif oldNode == node:
          print("oldNode:", oldNode.key)
          print("node:", node.key)
          print("Cycle occurs.")
          break

      else:
        newNode = oldNode

        if currentTable == self.T1:
          j = self.dhash(newNode.key)
          currentTable = self.T2
          currentTable[j], oldNode = newNode, currentTable[j]
        
        else:
          j = self.hash(newNode.key)
          currentTable = self.T1
          currentTable[j], oldNode = newNode, currentTable[j]
  
  def getList(self):
    T = self.T1 + self.T2
    T = [t for t in T if t != None]
    return T
```

```python
# example 1
epoches = 100
print("epoches:", epoches)

x = range(1, epoches + 1)

plt.figure(figsize=(10,8))

for k in range(1, 6):
  T = []
  for epoch in range(epoches):
    start_time = time.time()
    def heuristic(n):
              H_dist = {
                  'A': 10,
                  'B': 9,
                  'C': 999,
                  'D': 1,
                  'E': 6,
                  'G': 0,
              }
              return H_dist[n]

    A = Node('A', [2, 3], heuristic('A'))    # h value는 시작 노드, 도착 노드에 따라 바뀌므로 함수 사용
    B = Node('B', [2, 1, 9], heuristic('B'))
    C = Node('C', [1], heuristic('C'))
    D = Node('D', [6, 1], heuristic('D'))
    E = Node('E', [3, 6], heuristic('E'))
    G = Node('G', [9, 1], heuristic('G'))
    A.neighbors = [B, E]
    B.neighbors = [A, C, G]
    C.neighbors = [B]
    D.neighbors = [E, G]
    E.neighbors = [A, D]
    G.neighbors = [B, D]

    path, distance = aStarAlgo_Parallel(A, G, k)
    execution_time = time.time() - start_time
    T.append(execution_time)
  print("# of priority queues:", k)
  print("Shortest path:", path)
  print("Shortest distance:", distance)
  print(f"Mean execution time: {np.mean(T):.10f}\n")
  
  plt.plot(x, T, label=f"{k} priority queue(s)")
  time.sleep(0.5)

plt.xlabel("# of executions")
plt.ylabel("Time(s)          ", rotation=0)
plt.legend()
plt.show()
```
